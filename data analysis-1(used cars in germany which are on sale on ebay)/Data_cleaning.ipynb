```python
new_columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'no_of_pictures', 'postal_code',
       'last_seen']
```

```python
df.columns=new_columns
```

```python
df.head()
```

#### Convert dates to date type

```python
#change date_crawled, date_created, last_seen to date
df['date_crawled'] = pd.to_datetime(df['date_crawled']).dt.date
df['ad_created'] = pd.to_datetime(df['ad_created']).dt.date
df['last_seen'] = pd.to_datetime(df['last_seen']).dt.date
```

```python
df.head()
```

```python
df["no_of_pictures"].unique()
df.drop("no_of_pictures",axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df.fillna({"vehicle_type":"not mentioned","fuel_type":"unknown","unrepaired_damage":"unknown"},inplace=True)
```

```python
df.fillna({"gearbox":df["gearbox"].mode()[0],"model":df["model"].mode()[0]},inplace=True)
```
